# 3. Descriptive analysis with basic functions

In this section we will perform an initial descriptive analysis using basic functions.

```python
# Using the head function with a custom argument
df.head(30)
```

The output is showing the first 30 rows, because of the custom argument.

```python
# Using the tail() function to check last 5 rows of the dataframe
df.tail()
```

```python
# Checking the shape of the dataframe, with the shape function
df.shape
```

The shape function indicates the total number of rows and columns of the dataframe.

The info function gives informations about columns names and respective datatypes.

```python
# Using the dtypes function
df.dtypes
```

The dtypes function shows the datatype of all the dataframe columns.

```python
# Using the columns function
df.columns
```

# 4. Creating a list to import specific columns in a dataframe
** **

In this section we create a custom list of variables before importing the dataframe.

```python
# Creating a list containing only specific columns
vars_list = ['order_id', 'user_id', 'order_number', 'order_dow', 'order_hour_of_day', 'days_since_prior_order']
```

In the list, the eval_set column is not included

```python
# Importing the dataframe, using the list as an additional argument
df = pd.read_csv(os.path.join(path, '02. Data', 'Original Data', 'orders.csv'), usecols = vars_list)
```
